```python
import pandas as pd
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import TfidfTransformer
import json
import pickle
import numpy as np
```

```python
df = pd.read_table('../dataset/Export_agg_country_prod_yr.txt', encoding="latin-1",sep=',',dtype={'year':int,'HS':str,'SITC':str, 'CUCI': str})
```

```python
df = df[df["SITC"].notnull()]
df["value"] = df["value"].fillna(0)
```

```python
df["level"]=df.apply(lambda x: len(x["SITC"]),1)
```

### Lo hago a 4 dígitos

```python
df = df[df["level"]==4]
```

necesito una matriz de documentosXvocabulario --> paisXproducto

```python
matrix = df.pivot_table(index=['rep_iso','year'],columns='SITC',values='value')
matrix = matrix.fillna(0)
```

```python
matrix = matrix.to_sparse(fill_value=0)
```

## Utilizo TF-IDF para normalizar la importancia relativa

```python
tfidf = TfidfTransformer()

matrix = tfidf.fit_transform(matrix)
```

```python
matrix.shape
```

```python
### para hacer las pruebas de TF-IDF me quedo sólo con 30 componentes. 
```

```python
# #10 components
# lda_model_10 = LatentDirichletAllocation(n_components=10) #instance
# lda_model_10.fit(matrix) #fit
# pickle.dump(lda_model_10, open("lda10.pkl", 'wb')) #save
# #20 components
# lda_model_15 = LatentDirichletAllocation(n_components=15, max_iter=75, random_state=1234, learning_method='online',batch_size=200,learning_decay=0.5) #instance
# lda_model_15.fit(matrix) #fit
# pickle.dump(lda_model_15, open("lda15.pkl", 'wb')) #save

# #20 components
lda_model_20 = LatentDirichletAllocation(n_components=20, max_iter=75, random_state=1234,batch_size=200, learning_method='batch') #instance
lda_model_20.fit(matrix) #fit
pickle.dump(lda_model_20, open("lda20.pkl", 'wb')) #save
#30 components
lda_model_30 = LatentDirichletAllocation(n_components=30, max_iter=75, random_state=1234,batch_size=200, learning_method='batch') #instance
lda_model_30.fit(matrix) #fit
pickle.dump(lda_model_30, open("lda30.pkl", 'wb')) #save
# #50 components
# lda_model_50 = LatentDirichletAllocation(n_components=50) #instance
# lda_model_50.fit(matrix) #fit
# pickle.dump(lda_model_50, open("lda50.pkl", 'wb')) #save
```

```python
  
# load the model from disk
#lda_model = pickle.load(open(filename, 'rb'))
```

```python
productos = df.pivot_table(index=['rep_iso','year'],columns='SITC',values='value').columns
```

```python
lda_model_30.components_.shape
```

```python
product_df = pd.read_excel("data/UN Comtrade Commodity Classifications.xlsx")
```

```python
ClassS3 = product_df[product_df["Classification"]=='S3']
Codes = ClassS3.set_index('Code')['Description'].to_dict()
#Codes = ClassH0[['Code','Description']].to_dict(orient='index')
```

## 20 componentes

```python
norm_components = lda_model_20.components_ / lda_model_20.components_.sum(axis=1)[:, np.newaxis]
```

```python
for i in range(lda_model_20.components_.shape[0]):
    topic = norm_components[i]
    print("Componente %d:" % (i+1))
    top_n = [productos[j] for j in np.argsort(topic)[:-15 :-1]]
    top_n_props = topic[np.argsort(topic)][:-15 :-1]
    for j,producto in enumerate(top_n):
        try:
            print("{}-{}. Prop: {}%".format(producto, Codes[producto],+np.round(top_n_props[j]*100,3)))
        except:
            pass
    print("----------------------")
```

## Guardo los resultados para analizarlos en R

```python
components_df = pd.DataFrame(norm_components)
components_df.columns = [int(j) for j in productos]

components_df.to_csv("results/Dist_cadenas20.csv", index=False)
```

```python
result = lda_model_20.transform(matrix)
result = pd.DataFrame(result)
result.index = df.pivot_table(index=['rep_iso','year'],columns='SITC',values='value').index
result = result.reset_index()
result.to_csv("results/Dist_paises20.csv",index=False)
```
